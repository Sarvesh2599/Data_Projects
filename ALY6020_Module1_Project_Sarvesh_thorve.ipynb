```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('adult-all.csv')
```

```python
#Adding Colomn Names to the Dataframe
df.columns = ['age','workclass', 'fnlwgt', 'education', 'education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','salary']
df.head(50)
```

```python
#In this Dataset missing Data comes with "?". These values are replaces with NaN(Not a number)
df.replace("?", np.nan, inplace = True)
```

```python
#Detecting Missing Values
df.isnull().sum()
```

```python
#imputing missing values with mode
df = df.fillna(df.mode().iloc[0])
df.head(50)
```

```python
print(df.isnull().mean())
```

```python
count1 = (df['capital-loss'] == 0).sum()
count2 = (df['capital-gain'] == 0).sum()
print(count1)
print(count2)
```

```python
df.drop(['capital-loss', 'capital-gain','fnlwgt'], axis=1, inplace=True)
```

```python
plot1 = sns.countplot(df['salary'])
```

```python
def converter(salary):
    if salary == '<=50K':
        return 0
    else:
        return 1
```

```python
df['cluster'] = df['salary'].apply(converter)
```

```python
def gender(sex):
    if sex == 'Female':
        return 0
    else:
        return 1
```

```python
df['sex'] = df['sex'].apply(gender)
```

```python
sns.heatmap(df.corr(),vmin=-1, vmax=1, annot=True)
```

```python
dfa = df[['education-num','hours-per-week','age','sex']]
print('Total Row : ', len(dfa))
```

```python
sns.pairplot(df, hue='cluster')
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_dfa = StandardScaler()
dfa_std = sc_dfa.fit_transform(dfa.astype(float))
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=42).fit(dfa_std)
labels = kmeans.labels_
```

```python
new_dfa = pd.DataFrame(data = dfa_std, columns = ['education-num','hours-per-week','age','sex'])
new_dfa['label_kmeans'] = labels
```

```python
new_dfa
```

```python
#Plotting the graph
fig = plt.figure(figsize=(40,20))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(new_dfa.age[new_dfa.label_kmeans == 0], new_dfa["education-num"][new_dfa.label_kmeans == 0], new_dfa["hours-per-week"][new_dfa.label_kmeans == 0], c='blue', s=30, edgecolor='black')
ax.scatter(new_dfa.age[new_dfa.label_kmeans == 1], new_dfa["education-num"][new_dfa.label_kmeans == 1], new_dfa["hours-per-week"][new_dfa.label_kmeans == 1], c='red', s=30, edgecolor='black')
ax.scatter(new_dfa.age[new_dfa.label_kmeans == 2], new_dfa["education-num"][new_dfa.label_kmeans == 2], new_dfa["hours-per-week"][new_dfa.label_kmeans == 2], c='green', s=30, edgecolor='green')
ax.scatter(new_dfa.age[new_dfa.label_kmeans == 3], new_dfa["education-num"][new_dfa.label_kmeans == 3], new_dfa["hours-per-week"][new_dfa.label_kmeans == 3], c='orange', s=30, edgecolor='black')
centers = kmeans.cluster_centers_
ax.scatter(centers[:, 0], centers[:, 1], centers[:, 2], c='white', s=500);
plt.xlabel("education-num")
plt.ylabel("hours-per-week")
ax.set_zlabel('age')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split
```

```python
X = new_dfa
y = df['cluster']
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, y_train)
```

```python
pred = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, pred))
```

```python
print(classification_report(y_test, pred))
```

```python
error_rate = []
for i in range(1,8):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize = (10,6))
plt.plot(range(1,8),error_rate, color='blue',linestyle = '--', marker = 'o', markerfacecolor = 'red', markersize = 10)
plt.title('Error rate vs K')
plt.xlabel('K')
plt.ylabel("Error rate")
```

```python
knn = KNeighborsClassifier(n_neighbors = 4)
knn.fit(X_train, y_train)
pred = knn.predict(X_test)
print(confusion_matrix(y_test, pred))
print('\n')
print(classification_report(y_test, pred))
```
